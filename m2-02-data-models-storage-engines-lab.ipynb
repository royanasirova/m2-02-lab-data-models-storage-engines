### Task 2: Model the Same Data as Documents
Represent the same movie rental data using the document model.

Step 1: Create a list of customer documents (Python dictionaries) where each document contains the customer's information and a nested list of their rental history, with movie details embedded in each rental entry.

```python
customer_documents = []

for _, customer in customers.iterrows():
    customer_rentals = rentals[rentals["customer_id"] == customer["customer_id"]]
    rental_list = []
    for _, rental in customer_rentals.iterrows():
        movie = movies[movies["movie_id"] == rental["movie_id"]].iloc[0]
        rental_doc = {
            "rental_id": rental["rental_id"],
            "rental_date": rental["rental_date"],
            "return_date": rental["return_date"],
            "price": rental["price"],
            "movie": {
                "movie_id": movie["movie_id"],
                "title": movie["title"],
                "genre": movie["genre"],
                "release_year": movie["release_year"],
                "rating": movie["rating"]
            }
        }
        rental_list.append(rental_doc)
    customer_doc = {
        "customer_id": customer["customer_id"],
        "name": customer["name"],
        "email": customer["email"],
        "city": customer["city"],
        "signup_date": customer["signup_date"],
        "rentals": rental_list
    }
    customer_documents.append(customer_doc)
```

### Step 2: Execute the same four queries from Task 1 on the document collection:

Find all movies rented by a specific customer.

Find all movies rented by a specific customer.

```python
customer_id = 3

movies_rented = [
    rental["movie"]["title"]
    for customer in customer_documents
    if customer["customer_id"] == customer_id
    for rental in customer["rentals"]
]

movies_rented
```

Find the top 5 most-rented movies.

```python
movie_counts = {}
for customer in customer_documents:
    for rental in customer["rentals"]:
        title = rental["movie"]["title"]

        if title in movie_counts:
            movie_counts[title] += 1
        else:
            movie_counts[title] = 1
top5_movies = sorted(
    movie_counts.items(),
    key=lambda x: x[1],
    reverse=True
)[:5]
top5_movies
```

Compute total revenue per genre.

```python
revenue_per_genre = {}

for customer in customer_documents:
    for rental in customer["rentals"]:
        genre = rental["movie"]["genre"]
        price = rental["price"]

        if genre in revenue_per_genre:
            revenue_per_genre[genre] += price
        else:
            revenue_per_genre[genre] = price

revenue_per_genre
```

Find customers who have never rented a movie rated below 3.

```python
good_customers = []

for customer in customer_documents:
    ratings = [rental["movie"]["rating"] for rental in customer["rentals"]]
    if ratings and all(r >= 3 for r in ratings):
        good_customers.append(customer["name"])

good_customers
```

### Step 3: Write a markdown cell comparing the document queries to the relational queries. Which queries were easier to write? Which were harder? For which query pattern does the document model have a clear advantage?

In the relational model, data is stored in separate tables and queries
require joins between tables. This makes aggregation queries like
finding top movies or total revenue easier. In the document model, customer data and rental history are stored
together in one document. Queries about a single customer are easier
because no joins are needed. The document model is better for retrieving one customer's data,
while the relational model is better for analytical queries across
all customers and movies.

# Task 3: Model Relationships as a Graph
Build a graph representation of the rental data to answer relationship-based questions.

Step 1: Create a graph using Python dictionaries where:

Nodes represent customers, movies, and genres.
Edges represent relationships: rented (customer → movie), belongs_to (movie → genre), lives_in (customer → city).

```python
customers = [
    {"customer_id": 1, "name": "Alice", "city": "New York"},
    {"customer_id": 2, "name": "Bob", "city": "Los Angeles"},
    {"customer_id": 3, "name": "Charlie", "city": "New York"}
]

movies = [
    {"movie_id": 101, "title": "Matrix", "genre": "Sci-Fi"},
    {"movie_id": 102, "title": "Titanic", "genre": "Romance"},
    {"movie_id": 103, "title": "Inception", "genre": "Sci-Fi"}
]

rentals = [
    {"customer_id": 1, "movie_id": 101, "rating": 5},
    {"customer_id": 1, "movie_id": 102, "rating": 4},
    {"customer_id": 2, "movie_id": 103, "rating": 5},
    {"customer_id": 3, "movie_id": 101, "rating": 3}
]

graph = {
    "customers": {},  
    "movies": {},     
    "genres": set(),  
    "cities": set(),  
    "edges": {
        "rented": [],       
        "belongs_to": [],   
        "lives_in": []      
    }
}

for c in customers:
    graph["customers"][c["customer_id"]] = {"name": c["name"]}
    graph["cities"].add(c["city"])
    graph["edges"]["lives_in"].append((c["customer_id"], c["city"]))

for m in movies:
    graph["movies"][m["movie_id"]] = {"title": m["title"]}
    graph["genres"].add(m["genre"])
    graph["edges"]["belongs_to"].append((m["movie_id"], m["genre"]))

for r in rentals:
    graph["edges"]["rented"].append((r["customer_id"], r["movie_id"]))
    
import pprint
pprint.pprint(graph)
```

### Step 2: Write traversal functions to answer:

"Which movies did customer X rent?" — one hop from customer through rented edges.
"Which genres does customer X prefer?" — two hops: customer → movies → genres, then count.
"Which customers share a genre preference with customer X?" — three hops: customer X → movies → genres → other customers who rented movies in the same genre.
"Recommend movies for customer X" — find movies in their preferred genres that they have not yet rented.

```python
from collections import Counter
def movies_rented_by_customer(graph, customer_id):
    rented_edges = graph["edges"]["rented"]
    movie_ids = [movie_id for c_id, movie_id in rented_edges if c_id == customer_id]
    return [graph["movies"][m_id]["title"] for m_id in movie_ids]

def genres_preferred_by_customer(graph, customer_id):
    rented_edges = graph["edges"]["rented"]
    movie_ids = [movie_id for c_id, movie_id in rented_edges if c_id == customer_id]
    belongs_to = graph["edges"]["belongs_to"]
    genres = [genre for m_id, genre in belongs_to if m_id in movie_ids]
    return Counter(genres)

def customers_sharing_genre(graph, customer_id):
    preferred_genres = set(genres_preferred_by_customer(graph, customer_id).keys())
    rented_edges = graph["edges"]["rented"]
    belongs_to = graph["edges"]["belongs_to"]

    movie_to_genre = {m_id: genre for m_id, genre in belongs_to}
    genre_to_customers = {}
    for c_id, m_id in rented_edges:
        genre = movie_to_genre[m_id]
        genre_to_customers.setdefault(genre, set()).add(c_id)

    shared_customers = set()
    for genre in preferred_genres:
        shared_customers.update(genre_to_customers.get(genre, set()))

    shared_customers.discard(customer_id)
    return [graph["customers"][c]["name"] for c in shared_customers]

def recommend_movies(graph, customer_id):
    preferred_genres = set(genres_preferred_by_customer(graph, customer_id).keys())
    rented_movies = set(movies_rented_by_customer(graph, customer_id))
    
    movie_genre_map = {m_id: genre for m_id, genre in graph["edges"]["belongs_to"]}
    recommendations = [
        graph["movies"][m_id]["title"] 
        for m_id, genre in movie_genre_map.items() 
        if genre in preferred_genres and graph["movies"][m_id]["title"] not in rented_movies
    ]
    return recommendations
```

```python
customer_id = 1
print("Movies rented by Alice:", movies_rented_by_customer(graph, customer_id))
print("Alice's preferred genres:", genres_preferred_by_customer(graph, customer_id))
print("Customers sharing genre with Alice:", customers_sharing_genre(graph, customer_id))
print("Recommended movies for Alice:", recommend_movies(graph, customer_id))
```

### Step 3: Write a markdown cell explaining why the recommendation query (Task 3, question 4) would be difficult to express in SQL or with documents. What makes the graph model natural for this kind of traversal?

To recommend movies, we first find the movies a customer rented, then the genres of those movies, and finally other movies in the same genres that they haven’t rented yet. In SQL, this needs many table joins and filters, which can be hard to write and slow to run. In document models, we need to look inside nested documents, which is also tricky. In a graph, we can just follow the connections (edges) between customers, movies, and genres. This makes it easy to see relationships and find recommendations.

# Task 4: OLTP vs OLAP Workload Analysis
This task asks you to reason about how different database types handle different workloads.

Step 1: Create a DataFrame representing a log of 5,000 e-commerce transactions with columns: transaction_id, customer_id, product_id, amount, payment_method, timestamp, status (completed/refunded/pending).

```python
np.random.seed(42)
n = 5000
df = pd.DataFrame({
    "transaction_id": np.arange(1, n+1),
    "customer_id": np.random.randint(1, 1001, size=n),  
    "product_id": np.random.randint(1, 501, size=n),     
    "amount": np.round(np.random.uniform(5, 500, size=n), 2), 
    "payment_method": np.random.choice(["Credit Card", "PayPal", "Bank Transfer", "Gift Card"], size=n),
    "timestamp": pd.to_datetime(
        np.random.randint(
            pd.Timestamp("2026-01-01").value // 10**9, 
            pd.Timestamp("2026-03-01").value // 10**9,
            size=n
        ),
        unit='s'
    ),
    "status": np.random.choice(["completed", "refunded", "pending"], size=n, p=[0.8, 0.1, 0.1])
})
df.head()
```

### Step 2: Simulate OLTP-style operations:

Look up a single transaction by its ID (simulate a point query).

Update the status of a transaction from "pending" to "completed."
Check that the transaction is valid: the amount must be positive and the status must be one of the allowed values (simulate a consistency check).

```python
print("Lookup transaction 100:\n", df[df["transaction_id"] == 100])
```

```python
df = pd.concat([df, pd.DataFrame([{
    "transaction_id": df["transaction_id"].max() + 1,
    "customer_id": 1001,
    "product_id": 501,
    "amount": 250.75,
    "payment_method": "Credit Card",
    "timestamp": pd.Timestamp("2026-03-05 12:00:00"),
    "status": "completed"
}])], ignore_index=True)
```

```python
pending_txns = df[df["status"] == "pending"]
if not pending_txns.empty:
    txn_to_update = pending_txns.iloc[0]["transaction_id"]
    df.loc[df["transaction_id"] == txn_to_update, "status"] = "completed"
    print(f"\nUpdated transaction {txn_to_update} from pending to completed:")
    print(df[df["transaction_id"] == txn_to_update])
else:
    print("\nNo pending transactions to update.")
```

```python
valid_statuses = {"completed", "refunded", "pending"}
invalid = (df["amount"] <= 0) | (~df["status"].isin(valid_statuses))
print("\nInvalid transactions:\n", df[invalid] if invalid.any() else "All transactions are valid")
```

### Step 4: Write a markdown cell explaining: Why would OLTP operations typically use row-major storage while OLAP operations use column-major storage? How does the access pattern of each workload type map to the memory layout that serves it best?

OLTP usually reads or updates one row at a time, so storing data row by row is faster because all fields of a row are together. OLAP usually analyzes a few columns across many rows, so storing data column by column is faster because only the needed columns are read. In short, row storage works best for single-row operations and column storage works best for large-scale analysis.

For each of the following real-world scenarios, recommend a data model (relational, document, or graph) and a processing paradigm (OLTP or OLAP). Write 3-4 sentences justifying each choice.

1.A hospital patient record system where doctors need to quickly look up a patient's complete medical history.
2.A social network that needs to suggest "friends of friends" connections.
3.A financial reporting system that computes monthly revenue breakdowns across product categories and regions.
4.An IoT platform that receives sensor readings from 10,000 devices and needs to detect anomalies in real time.
5.A content management system where each article has a different set of metadata fields.

1. A relational model with OLTP is best because patient data is structured and doctors need fast access to complete records. Row storage allows quick lookups and safe updates. Transactions ensure consistency when records are modified.
2. A graph model works best since friendships are naturally nodes and edges. Multi-hop queries like “friends of friends” are fast and easy. This also supports dynamic updates as users and relationships grow.
3. A relational or columnar model with OLAP is ideal for aggregating revenue by product and region. Column-oriented storage speeds up computations on large datasets. OLAP is optimized for analyzing many rows efficiently.
4. A document or time-series model with OLTP-style streaming fits because devices send frequent readings with variable fields. Flexible documents allow storing different data structures. Fast inserts and queries enable real-time anomaly detection.
5. A document model with OLTP is best since articles have different metadata fields. Flexible documents handle diverse structures naturally. OLTP allows fast updates and retrievals of individual articles.
